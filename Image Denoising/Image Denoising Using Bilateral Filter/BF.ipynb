```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_denoising(image):
    # Convert the image to float32 for bilateral filtering
    image = np.float32(image)

    # Apply bilateral filter for image denoising
    denoised_image = cv2.bilateralFilter(image, -1, 25, 25)

    # Convert the denoised image back to uint8
    denoised_image = np.uint8(denoised_image)

    return denoised_image

# Load the noisy image
image = cv2.imread("Robot.jpg", 0)

# Perform image denoising
denoised_image = image_denoising(image)

# Display the original and denoised images using Matplotlib
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Noisy Image")
axes[0].axis("off")
axes[1].imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Denoised Image")
axes[1].axis("off")
plt.show() 
```
