```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an image
src = cv2.imread("02.jpg")

# Convert the image to grayscale
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# Apply median blur to reduce noise
gray = cv2.medianBlur(gray, 5)

# Detect circles using Hough transform
rows = gray.shape[0]
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, rows / 8,
                           param1=100, param2=30,
                           minRadius=60, maxRadius=100)

# Draw detected circles on the original image
if circles is not None:
    circles = np.uint16(np.around(circles))
    for c in circles[0, :]:
        center_x, center_y, radius = c
        center = (center_x, center_y)
        # Draw the center of the circle
        cv2.circle(src, center, 1, (0, 100, 100), 3)  # Center
        # Draw the outline of the circle
        cv2.circle(src, center, radius, (255, 0, 255), 3)  # Outline

# Display the image with detected circles
plt.imshow(src[..., ::-1])
plt.title("Image with Detected Coins")
plt.show()

# Print the number of detected coins
if circles is not None:
    print("Number of coins:", len(circles[0, :]))
else:
    print("No coins detected.")
```
